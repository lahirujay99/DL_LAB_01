```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Task 1.1 : Create a Random 4x4 Array from an Exponential Distribution

array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 array from exponential distribution:\n", array_4x4)
```

```python
# Task 1.2 : Create a Random 100000x1 Array and Visualize

data_exp = np.random.exponential(scale=1.0, size=100000)
data_uniform = np.random.rand(100000)
data_normal = np.random.randn(100000)

plt.hist(data_exp, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(data_uniform, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(data_normal, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Task 1.3 : 3D Plot of Z = X^2 + Y^2

from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

plt.title("3D Plot of Z = X^2 + Y^2")
plt.show()
```

```python
# Task 1.4 : Pearson and Spearman Correlation

# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Adjust the path to the specific folder and file in your Google Drive
file_path = '/content/drive/My Drive/Colab Files - DL/Au_nanoparticle_dataset.csv'

# Load the dataset
data = pd.read_csv(file_path)

# Task 1.4: Pearson and Spearman Correlation
corr_pearson = data.corr(method='pearson')
corr_spearman = data.corr(method='spearman')

fig, ax = plt.subplots(1, 2, figsize=(15, 5))

sns.heatmap(corr_pearson, annot=True, ax=ax[0], cmap='coolwarm')
ax[0].set_title('Pearson Correlation')

sns.heatmap(corr_spearman, annot=True, ax=ax[1], cmap='coolwarm')
ax[1].set_title('Spearman Correlation')

plt.show()
```
