```python
# Task 2 : Execute the Given Tasks

# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Adjust the path to the specific folder and file in your Google Drive
file_path = '/content/drive/My Drive/Colab Files - DL/Au_nanoparticle_dataset.csv'

# Load the dataset
df = pd.read_csv(file_path)
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
# Task 2.2 : Display the First 20 Samples

# Display the first 20 samples
print(new_df.head(20))
```

```python
# Task 2.3 : Calculate Mean, Standard Deviation, and Quartiles

# Calculate mean, standard deviation, and quartiles
mean = new_df.mean()
std_dev = new_df.std()
quartiles = new_df.quantile([0.25, 0.5, 0.75])

print("Mean:\n", mean)
print("Standard Deviation:\n", std_dev)
print("Quartiles:\n", quartiles)
```

```python
# Task 2.4 : Plot Histograms in a 1x4 Layout

# Plot histograms in a 1x4 layout
fig, axs = plt.subplots(1, 4, figsize=(20, 5))
new_df.hist(ax=axs)
plt.show()
```

```python
# Task 2.5 : Visualize Scatter Plots and Histograms Using Pairplot

# Visualize scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()
```

```python
# Task 2.6 : Modify PairGrid for Different Plots

# Modify PairGrid for different plots
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot)  # Bivariate histogram
g.map_diag(sns.histplot, kde=True)  # Histogram with KDE on the diagonal
g.map_lower(sns.kdeplot)  # Bivariate KDE
plt.show()
```
